```python
import statsmodels.formula.api as smf
```

```python
model=smf.ols("Mort~Lat",data=skincancer).fit()
```

```python
print(model.summary())
```

```python
model.params
```

```python
pred=model.predict(skincancer.iloc[:,0])
pred
```

```python
plt.scatter(skincancer['Lat'],skincancer['Mort'],color='r')
plt.plot(skincancer['Lat'],pred,color='b')
plt.xlabel('Latitude')
plt.ylabel('Mortality Rate')
```

```python
#with the help of this we can able to predict that increase in latitude decrease in death rate.
```
