```python
import pandas as pd
import numpy as np  #loading libraries
```

```python
##here there are only numeric values from 1st column to last column so extracting these columns from the dataset
univ=University.iloc[:,1:]
univ
```

```python
#if we observe the above data few features have higher value and few features have lower values so we need to scale it
from sklearn.preprocessing import scale
df_norm=scale(univ)
df_norm
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=6)  #here there are six features so i took n_components as 6
pca_values=pca.fit_transform(df_norm)
pca_values
```

```python
#the below correlation are used for just explantion of concept 
```

```python
univ.corr() # finds correlation after standardizing
```

```python
univ.cov() #finds correlation before standardizing the data
```

```python
# The amount of variance that each PCA explains is 
var=pca.explained_variance_ratio_
var
pca.components_
```

```python
pca.components_[0]   #first principal component
```

```python
var
```

```python
# Cumulative variance 
var1= np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
import matplotlib.pyplot as plt
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```
