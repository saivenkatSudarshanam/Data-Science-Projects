```python
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup as bs
from wordcloud import WordCloud
import re
```

```python
LG_ACReviews = []
```

```python
for i in range(1,11):
    lg=[]
    url = "https://www.amazon.in/LG-Convertible-Anti-Virus-Protection-MS-Q18YNZA/product-reviews/B08R918D6Y/ref=cm_cr_arp_d_viewpnt_lft?ie=UTF8&reviewerType=all_reviews&filterByStar=positive&pageNumber="+str(i)
    response=requests.get(url)
    soup = bs(response.content,"html.parser")
    review = soup.findAll("span",attrs ={"class","a-size-base review-text review-text-content"})
    for i in range(len(review)):
        lg.append(review[i].text)
    LG_ACReviews=LG_ACReviews+lg

with open("lg.txt","w",encoding='utf8') as output:
    for i in LG_ACReviews:
        output.write(i+"\n")
```

```python
lg_rev_string = " ".join(LG_ACReviews)
```

```python
lg_rev_string
```

```python
lg_rev_string = re.sub("[^A-Za-z" "]+"," ",lg_rev_string).lower()
lg_rev_string=re.sub("[0-9" "]+"," ",lg_rev_string)
```

```python
lg_rev_string
```

```python
lg_rev_words = lg_rev_string.split(" ")
lg_rev_words
```

```python
with open("E:\Top Mentor\stop.txt",'r') as sw:
    stopwords = sw.read()
stopwords=stopwords.split('\n')
stopwords
```

```python
lg_rev_words=[w for w in lg_rev_words if not w in stopwords]
lg_rev_words
```

```python
lg_rev_string=" ".join(lg_rev_words)
lg_rev_string
```

```python
word_cloud_lg =WordCloud(width=5000,height=2500,background_color='black').generate(lg_rev_string)
plt.imshow(word_cloud_lg)
```
