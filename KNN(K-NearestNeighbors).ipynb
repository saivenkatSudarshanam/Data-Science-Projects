```python
x=cancer.iloc[:,1:]
y=cancer['diagnosis']
```

```python
x
```

```python
y
```

```python
#Normalizing the data because it helps in distance.a variable that ranges between 0 and 100 will outweigh a variable that ranges between 0 and 1. Using these variables without standardization in effect gives the variable with the larger range a bigger weight in the analysis
```

```python
from sklearn.preprocessing import StandardScaler
s_scalar=StandardScaler()
x_standards=s_scalar.fit_transform(x)
x_standards
```

```python
#scaling is done now we have to split data into train and test data set
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_standards,y,test_size=0.2,random_state=66)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
print("training dataset:",x_train.shape,y_train.shape)
print("testing dataset:",x_test.shape,y_test.shape)
```

```python
#choosing k value and applying with the help of knn algo
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
#checking score of test and train data
knn.score(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
#predicting values
```

```python
pred=knn.predict(x_train)
pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
#checking other parameters like confusion matrix classification report and accuracy
print("\n K-nearest neighbours-Confusion matrix on training dataset:\n\n",pd.crosstab(y_train,pred,rownames=['Actual'],colnames=['predict']))
```

```python
print("KNN Accuracy score of training dataset:",round(accuracy_score(y_train,pred),3))
```

```python
print("knn neighbours Classification report of training dataset:\n",classification_report(y_train,pred))
```

```python
pred1=knn.predict(x_test)
pred1
```

```python
#now checking accuracy confusion matrix and classification report on testing data set
```

```python
print("\n Knn neighbours -confusion matrix on testing dataset:\n\n",pd.crosstab(y_test,pred1,rownames=['Actual'],colnames=['Predict']))
```

```python
print("knn neighbours -accuracy score on testing dataset:",round(accuracy_score(y_test,pred1),3))
```

```python
print("\nknn neighbours- Classification report on testing dataset:\n\n",classification_report(y_test,pred1))
```

```python
#try KNN for diffrent k nearest neighbor from 1 to 15
training_accuracy=[]
testing_accuracy=[]
neighbors_count=range(1,15)
for n_neighbors in neighbors_count:
    knn=KNeighborsClassifier(n_neighbors=n_neighbors)
    knn.fit(x_train,y_train)
    training_accuracy.append(accuracy_score(y_train,knn.predict(x_train)))
    testing_accuracy.append(accuracy_score(y_test,knn.predict(x_test)))
plt.plot(neighbors_count,training_accuracy,label='Accuracy of training dataset')
plt.plot(neighbors_count,testing_accuracy,label='Accuracy of testing dataset')
plt.xlabel('KNeighbors')
plt.ylabel('accuracy')
plt.legend()
plt.show()
    
```
