```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer #importing bagofwords and tfidf libraries
```

```python
#loading dataset
email=pd.read_csv("E:\SaiVenkat\ham_spam.csv",encoding='ISO-8859-1')
email
```

```python
#problem statement is with the help of text mail we have to identify whther the mail is ham or spam
```

```python
import re #importing re for string pattern matching
stop_words=[]
with open("E:\SaiVenkat\stop.txt") as f:
    stop_words=f.read()
```

```python
#loading stop words file just to see what  are the stop words present in it
stop_words
```

```python
stop_words=stop_words.split("\n") #splitting stop words with \n 
stop_words
```

```python
#defining fucntion for cleaning text
def cleaning_text(i):
    i=re.sub("[^A-Za-z " "]"," ",i).lower()
    i=re.sub("[0-9 " "]"," ",i)
    w=[]
    for word in i.split(" "):
        if len(word)>3:
            w.append(word)
    return (" ".join(w))
```

```python
#cleanining text from email data
email['text']=email.text.apply(cleaning_text)
email.text
```

```python
email #everything looks good now we have filtereed only text data from the text column
```

```python
#its time to identify what are the not null rows in it
email=email.loc[email.text!= " ", : ]
email
```

```python
#defining function for splitting words
def split_into_words(i):
    return [word for word in i.split(" ")]
```

```python
#Count vectorization for bagof words or in normal words splitting the words and checking their frequency
email_bagofwords=CountVectorizer(analyzer=split_into_words).fit(email.text)
```

```python
email_bagofwords
```

```python
email_bagofwords.vocabulary_ #checking the words and their frequencies
```

```python
#splitting data for train test
```

```python
from sklearn.model_selection import train_test_split
email_train,email_test=train_test_split(email,test_size=0.3)
```

```python
#train data
email_train
```

```python
#testdata
email_test
```

```python
#in the above we just fitted text column data now we need to transform that text data into normalised data
#for all messages i.e for complete texts in email.text
all_email_matrix=email_bagofwords.transform(email.text)
all_email_matrix
```

```python
#for train messages
train_email_matrix=email_bagofwords.transform(email_train.text)
train_email_matrix
```

```python
#for test messages
test_email_matrix=email_bagofwords.transform(email_test.text)
test_email_matrix
```

```python
#now without Tfidf matrix perform modeling with mb and gb
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
```

```python
#fitting model
clf_MB=MultinomialNB()
clf_MB.fit(train_email_matrix,email_train.type) #train_email_matrix is normalised x data and type is y data
```

```python
train_pred_MB=clf_MB.predict(train_email_matrix) #predicting y values with the help of x train data
train_pred_MB
```

```python
accuracy_train_MB=clf_MB.score(train_email_matrix,email_train.type) #two ways to calculate accuarcy 
accuracy_train_MB
```

```python
accuracy_train_MB1=np.mean(train_pred_MB==email_train.type)
accuracy_train_MB1
```

```python
#now calculating testing data 
```

```python
test_pred_MB=clf_MB.predict(test_email_matrix) #prediciting testing data
test_pred_MB
```

```python
accuracy_test_MB=clf_MB.score(test_email_matrix,email_test.type)
accuracy_test_MB
```

```python
accuracy_test_MB1=np.mean(test_pred_MB==email_test.type)
accuracy_test_MB1
```

```python
#done with MB now with GB
clf_GB=GaussianNB()
clf_GB.fit(train_email_matrix.toarray(),email_train.type.values)
```

```python
#prediciting and finding accuracy of training data
train_pred_GB=clf_GB.predict(train_email_matrix.toarray())
train_pred_GB
```

```python
accuracy_train_GB=clf_GB.score(train_email_matrix.toarray(),email_train.type.values) #both the ways are correct
accuracy_train_GB
```

```python
accuracy_train_GB1=np.mean(train_pred_GB==email_train.type.values)
accuracy_train_GB1
```

```python
#predicting for testing data and accuracy score
test_pred_GB=clf_GB.predict(test_email_matrix.toarray())
test_pred_GB
```

```python
accuracy_test_GB=clf_GB.score(test_email_matrix.toarray(),email_test.type.values)
accuracy_test_GB
```
