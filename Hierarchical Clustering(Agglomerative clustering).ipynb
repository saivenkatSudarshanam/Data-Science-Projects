```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading dataset
```

```python
univ=pd.read_csv('E:\SaiVenkat/Universities.csv')
univ
```

```python
#checking whether there are any missing values present in or not
univ.info()
```

```python
def norm_func(i):   #defining normal function
    x=(i-i.mean())/i.std()
    return (x)
```

```python
df_norm=norm_func(univ.iloc[:,1:])
df_norm
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch  #for creating a dendogram
```

```python
z=linkage(df_norm,method='complete',metric='euclidean')
z
```

```python
plt.figure(figsize=(15,5));
plt.title('Hierarchical clustering')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,leaf_font_size=8.,leaf_rotation=0,)
plt.show()
```

```python
#choosing the clusters with the help of alggometric clustering
from sklearn.cluster import AgglomerativeClustering
h_complete=AgglomerativeClustering(n_clusters=3,linkage='complete',affinity='euclidean')
h_complete.fit(df_norm)
```

```python
h_complete.labels_
```

```python
univ['cluster']=h_complete.labels_
univ
```
