```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense,Activation,Layer,Lambda
from sklearn.model_selection import train_test_split
#imported all the necessary libraries
```

```python
#loading the dataset
Concrete=pd.read_csv("E:\SaiVenkat\concrete (1).csv") 
Concrete
```

```python
#In above dataset features are cement,slag,ash,water,superplastic,coarseagg,fineagg,age and label is strength
```

```python
Concrete_model=Sequential()  #model building
Concrete_model.add(Dense(50,input_dim=8,activation='relu'))  #hidden layers
Concrete_model.add(Dense(40,activation='relu'))
Concrete_model.add(Dense(20,activation='relu'))
Concrete_model.add(Dense(10,activation='relu'))
Concrete_model.add(Dense(1,kernel_initializer="normal"))  
Concrete_model.compile(loss="mean_squared_error",optimizer="adam",metrics=['mse']) #compiling
```

```python
Column_names=list(Concrete.columns)
predictor=Column_names[0:8]   #spliting features and target columns
target=Column_names[8]
```

```python
First_model=Concrete_model
First_model.fit(np.array(Concrete[predictor]),np.array(Concrete[target]),epochs=1000)
pred_train=First_model.predict(np.array(Concrete[predictor]))
pred_train = pd.Series([i[0] for i in pred_train])
rmse_value = np.sqrt(np.mean((pred_train-Concrete[target])**2))
```
